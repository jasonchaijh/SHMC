```python
import math
import torch
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
from epde import ePDE
```

```python
import scipy.io as sio
y=sio.loadmat('u-3.mat')
#y=np.flipud(u3['u3'])
y=y['u2'].flatten('F').reshape([-1,1])
y=torch.tensor(y.copy())
x=np.meshgrid(np.linspace(0,1,51),np.linspace(0,1,51))
x=torch.tensor(np.c_[x[0].T.flatten(),x[1].T.flatten()].astype(np.float)).T

#choose some points
x=x[:,40:2000:200]
y=y[40:2000:200]
epde=ePDE(d=6,x=x,y=y,sigma2_y=1e-2,l=0.2,nsim=30,stepsize=0.05)
L=5
epi=0.1
iteration=1000
for hmc_run in range(iteration):
    l=np.random.randint(L)
    epde.hmc(l,epi)
#epde.thetalist
```

```python
print(epde.thetalist.shape)
#thetalist=epde.thetalist
plt.title('')
plt.scatter(epde.thetalist[0],epde.thetalist[1],10)
```

```python
fig,ax=plt.subplots(6,6,figsize=(10,10))

for i in range(6):
    for j in range(i+1,6):
        fig.delaxes(ax[i][j])
for i in range(1,6):
    for j in range(1,i+1):
        ax[i][j].yaxis.set_visible(False)
ax[0][0].yaxis.set_visible(False)
for i in range(0,5):
    for j in range(0,i+1):
        ax[i][j].xaxis.set_visible(False)
        
ax[1][0].scatter(epde.thetalist[0],epde.thetalist[1],1)
ax[2][0].scatter(epde.thetalist[0],epde.thetalist[2],1)
ax[3][0].scatter(epde.thetalist[0],epde.thetalist[3],1)
ax[4][0].scatter(epde.thetalist[0],epde.thetalist[4],1)
ax[5][0].scatter(epde.thetalist[0],epde.thetalist[5],1)

ax[2][1].scatter(epde.thetalist[1],epde.thetalist[2],1)
ax[3][1].scatter(epde.thetalist[1],epde.thetalist[3],1)
ax[4][1].scatter(epde.thetalist[1],epde.thetalist[4],1)
ax[5][1].scatter(epde.thetalist[1],epde.thetalist[5],1)

ax[3][2].scatter(epde.thetalist[2],epde.thetalist[3],1)
ax[4][2].scatter(epde.thetalist[2],epde.thetalist[4],1)
ax[5][2].scatter(epde.thetalist[2],epde.thetalist[5],1)

ax[4][3].scatter(epde.thetalist[3],epde.thetalist[4],1)
ax[5][3].scatter(epde.thetalist[3],epde.thetalist[5],1)
ax[5][4].scatter(epde.thetalist[4],epde.thetalist[5],1)

ax[0][0].set_title(r'$\theta_1$')
ax[1][1].set_title(r'$\theta_2$')
ax[2][2].set_title(r'$\theta_3$')
ax[3][3].set_title(r'$\theta_4$')
ax[4][4].set_title(r'$\theta_5$')
ax[5][5].set_title(r'$\theta_6$')

for i in range(6):
    s = epde.thetalist[i]
    p, x = np.histogram(s, bins=100) # bin it into n = N//10 bins
    x = x[:-1] + (x[1] - x[0])/2   # convert bin edges to centers
    f = UnivariateSpline(x, p, s=1800)
    ax[i][i].plot(x,f(x))


plt.suptitle('One- and two-dimensional posterior marginals of '+ r'$\theta_1,\theta_2,\theta_3,\theta_4,\theta_5,\theta_6$'\
             +' using stochastic HMC',y=0)
plt.savefig('shmc.png',dpi=300)
plt.show()
```
